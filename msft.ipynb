```python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

api_key = os.environ['APIKEY']
```

```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_msft = pdr.get_data_tiingo('MSFT', api_key=api_key)
```

```python
df_msft
```

```python
df_msft['high'].plot(figsize=(12,4))
```

```python
df_msft.index
```

```python
df_msft.index.set_levels(pd.to_datetime(df_msft.index.get_level_values(1)).date, level=1, inplace=True)
```

```python
df_msft.index
```

```python
df_msft = df_msft.droplevel('symbol', axis=0)
```

```python
df_msft.index
```

```python
df_msft.index = pd.to_datetime(df_msft.index)
```

```python
df_msft.index
```

```python
index=df_msft.loc['2022-01-01':'2023-10-01'].index
share_open=df_msft.loc['2022-01-01':'2023-10-01']['open']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure,axis=plt.subplots()
plt.tight_layout()
## Preventing overlapping
figure.autofmt_xdate()
axis.plot(index,share_open)
```

```python
from datetime import datetime
```

```python
# time resampling
df_msft.resample(rule='A').min()
```

```python
df_msft.resample(rule='A').min()['open'].plot()
```

```python
#Quarterly frequency
df_msft.resample(rule='QS').min()
```

```python
#Business end frequency
df_msft.resample(rule='BA').min()
```

```python
df_msft['open'].resample(rule = 'A').mean().plot(kind = 'bar')
```

```python
df_msft['open'].resample(rule = 'M').mean().plot(kind = 'bar', figsize = (16, 5))
```
